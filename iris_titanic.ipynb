```python
# prompt: cria clusters com as novas colunas petal width (normalized) e petal width (standardized)

# Criando o modelo KMeans para a coluna normalizada
kmeans_norm = KMeans(n_clusters=3)
kmeans_norm.fit(df_iris[['petal width (normalized)']])

# Adicionando as labels dos clusters aos dados originais para a coluna normalizada
df_iris['cluster_normalized'] = kmeans_norm.labels_

# Criando o modelo KMeans para a coluna padronizada
kmeans_padr = KMeans(n_clusters=3)
kmeans_padr.fit(df_iris[['petal width (standardized)']])

# Adicionando as labels dos clusters aos dados originais para a coluna padronizada
df_iris['cluster_standardized'] = kmeans_padr.labels_

# Exibindo o DataFrame com as novas colunas de clusters
df_iris
```

```python
# prompt: gerar o grafico dos clusters das colunas cluster_normalized e cluster_standardized

import plotly.express as px

# Gráfico para a coluna normalizada
fig_norm = px.scatter(df_iris, x='petal length (cm)', y='sepal length (cm)', color='cluster_normalized',
                     title='Clusters da Coluna Normalizada')
fig_norm.show()

# Gráfico para a coluna padronizada
fig_padr = px.scatter(df_iris, x='petal length (cm)', y='sepal length (cm)', color='cluster_standardized',
                     title='Clusters da Coluna Padronizada')
fig_padr.show()
```
